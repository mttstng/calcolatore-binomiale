```python

import numpy as np # linear algebra

import matplotlib as mpl 

import matplotlib.pyplot as plt

import plotly.express as px

import scipy.special

import pandas as pd
```

```python
print("Scegli il numero di prove dell'esperimento, e premi invio")
N= int(input()) #numero di prove
print("\n\nIl numero di prove scelto è "+str(N))


while True:
    p= float(input("\nScegli il valore di probabilità di riuscita dell'esperimento e premi invio:")) #probabilità di successo
    
    if not 0<= p <=1:
        print("\n\nLa probabilità è un numero tra zero e uno, tenta di nuovo!")
        continue
    else:
        print("\n\nLa probabilità di riuscita dell'esperimento è "+str(p)+"\n\n")
        break



n_p= 1-p #probabilità di non successo



C_N_k = []

for k in np.arange(N+1):
    C_N_k.append(scipy.special.binom(N, k)*(p**k)*(n_p**(N-k)))
    

df = pd.DataFrame(
{'numero_di_successi_k' :np.arange(N+1),
'probabilità' : C_N_k},
index = np.arange(N+1))

with pd.option_context('display.max_rows', None, 'display.max_columns', df.shape[1]):   # display(df)
    display(df)




fig = px.bar(df, x="numero_di_successi_k", y="probabilità",
             title= r'$\text{Calcolatore binomiale: }\ P_n(k)=\binom{n}{k} p^k (1-p)^{n-k}$ ',
             color = "probabilità",
             color_continuous_scale= px.colors.sequential.matter,
             template="presentation",
             
            )

fig.update_layout(
    xaxis = dict(
        tickmode = 'linear',
        tick0 = 0,
        dtick = 1
    )
)

valori= f"numero di prove n={str(N)} \nprobabilità di successo p={str(p)}"
valori_C= f"n={str(N)} <br> p={str(p)}"

print(valori)

y_lim= np.amax(C_N_k) 
fig.add_annotation(x=0.98*N, y=1.3*y_lim,
            text=valori_C,
            showarrow=False,
            yshift=10,
            bgcolor='green',
            hovertext = valori,
                  )



fig.show()
```
